```python
#import libraries
import matplotlib.pyplot as plt
import numpy as np

#import datasets
from sklearn import datasets
```

```python
# plot the Iris data in a scatter diagram
def plotData(data, labels,fName,col1,col2, centroids = np.empty([0,0,0,0])):
    #plot the titles
    plt.title("The Iris Dataset  labels")
    plt.xlabel(fNames[col1])
    plt.ylabel(fNames[col2])
    # plot the data with different color for every label and size of 50 pixels
    myplot =plt.scatter(data[:,col1],data[:,col2],c = labels, s=50)

    # plot the legend
    plt.legend(handles=myplot.legend_elements()[0], labels=list(set(labels)))

    # plot the centroids if exists in blue and with + marker
    if (centroids.any()):
        plt.scatter(centroids[:,col1], centroids[:,col2], c='blue', marker = "+", s=150)
    plt.show(myplot)
```

```python
## Example using the plotData function

import numpy as np
import matplotlib.pyplot as plt

# fake 2-D points
data = np.array([
    [1.0, 2.0],
    [1.2, 1.8],
    [3.5, 4.0],
    [3.7, 3.8],
    [6.0, 2.0],
    [6.2, 1.9]
])

labels = np.array([0,0,1,1,2,2])   # three groups

fNames = ["featA","featB"]        # names used on axes

# no centroids
plotData(data, labels, fNames, 0, 1)

```

```python
# plot labels
# plot a bar diagram for the labels
def plotLabels(labels):
    xmax=len(labels)
    plt.title("The Iris Dataset labels")
    plt.xlabel("Iris sample number")
    plt.ylabel("label")
    plt.bar(range(xmax),labels)
    plt.show()
```

```python
## Example using the plotLabels function

import numpy as np

# labels for 6 samples
labs = np.array([0,0,1,1,2,2])

plotLabels(labs)
```

```python
# vetor distance
# Compute the distance between vectors
def vecDist(vec1,vec2):
    vecDiff = vec1-vec2
#   compute the distance ("pitagoras")
    vecSqr =  np.square(vecDiff)
    vecSum =  np.sum(vecSqr)
    return np.sqrt(vecSum)
```

```python
# vector centroids
# for every label compute the average of every data colomn
def vecCentroids(data,labels):
    # number of different labels
    nLabels=len(set(labels))
    # number of data features - colmons
    nCols = data.shape[1]
    inLabels = labels-1
    # compute sum and count of every colomn for every label
    sum = np.zeros((nLabels,nCols))
    count = np.zeros((nLabels,nCols))
    for i,elem in enumerate(data):
        sum[inLabels[i]]+=elem
        count[inLabels[i]]+=1
    return(sum/count)
```
